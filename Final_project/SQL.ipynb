Определим для каждой книги количество обзоров и среднюю оценку 

```python
query = '''
SELECT
    books.title AS title,
    books.book_id AS id,
    AVG(ratings.rating) AS avg_rating,
    COUNT(DISTINCT reviews.review_id) AS cnt_review
FROM
    books
INNER JOIN ratings ON ratings.book_id = books.book_id
INNER JOIN reviews ON reviews.book_id = books.book_id
GROUP BY 
    books.title,
    books.book_id
ORDER BY
    cnt_review DESC
limit 10
'''
pd.io.sql.read_sql(query, con = engine)
```

## Вывод:

В Результате запроса мы вывели список книг c числом обзоров и оценкой. Чтобы таблица была не слишком большая, мы вывели первые 10 результатов.

Определим издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры. 

```python
query = '''
SELECT
    publishers.publisher AS publisher_name,
    COUNT(books.book_id) AS cnt_books
FROM
    books
INNER JOIN publishers ON publishers.publisher_id = books.publisher_id
GROUP BY 
    publishers.publisher
HAVING
    SUM(books.num_pages) > 50
ORDER BY
    cnt_books DESC
limit 1
'''
pd.io.sql.read_sql(query, con = engine)
```

## Вывод:
    
В Результате запроса мы определили, что издательство Penguin Books выпускает больше всего книг, толщиной более 50 страниц.

Определим автора с самой высокой средней оценкой книг — при учете только книг с 50 и более оценками

```python
query = '''
SELECT 
    SUBQ.author AS author, 
    AVG(SUBQ.avg_rating) AS avg_rating
FROM 
    (SELECT
        authors.author as author,
        books.book_id AS id,
        AVG(ratings.rating) AS avg_rating,
        COUNT(ratings.rating) AS cnt_rating
FROM 
    authors
INNER JOIN books ON books.author_id = authors.author_id
INNER JOIN ratings ON ratings.book_id = books.book_id
GROUP BY 
    author,
    id
HAVING
    COUNT(ratings.rating) >= 50) AS SUBQ
GROUP BY
    author
ORDER BY
    avg_rating DESC
limit 10
'''
pd.io.sql.read_sql(query, con = engine)
```

## Вывод:
    
Автором с самой высокой средней оценкой книг является - J.K. Rowling/Mary GrandPré

Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query = '''
SELECT
    AVG(SUBQ.count) AS mean
FROM
    (SELECT COUNT(*)
        FROM reviews
            WHERE username in
                (SELECT username
                    FROM ratings
                    GROUP BY username
                    HAVING COUNT(rating) > 50)
            GROUP BY username)
            AS SUBQ
'''
pd.io.sql.read_sql(query, con = engine)
```

## Вывод:
    
Среднее количество обзоров от пользователей, которые поставили больше 50 оценокв - 24.33
