Начнем с высоты пололков и посомтрим на их количество, и средние параметры с медианой.

```python
df['ceiling_height'].value_counts()
```

Как мы видим, cреднее и медиана находятся где-то между 2.7. Если обращать внимание на строительные нормы на 2021 год, то как раз высота потолков 2.7 является железным стандартом. 

Мне кажется что можно прикинуть высоту потолков в зависимости от этажности и от этого произвести замену.

```python
def howlevel(num):
    if num >= 1 and num <= 5:
        return 'малоэтажки'
    elif num > 5 and num < 10:
        return 'многоэтажки'
    else:
        return 'высокоэтажки'

df['new_floor'] = df['floors_total'].apply(howlevel)
```

```python
k = []
for i in ['малоэтажки', 'многоэтажки', 'высокоэтажки']:
    result = df[df['new_floor'] == i]['ceiling_height'].median()
    k.append(result)

level = pd.DataFrame(data = k, columns = ['level'])
```

```python
level
```

Теперь произведем замену пропусков в зависимости от этажности

```python
columns = ['малоэтажки', 'многоэтажки', 'высокоэтажки']
for i in range(len(columns)):
    df.loc[df['new_floor'] == columns[i],'ceiling_height'] = df.loc[df['new_floor'] == columns[i],'ceiling_height'].fillna(level['level'][i])
```

Теперь посмотрим, все ли удалось изменить. 

```python
df['ceiling_height'].isna().sum()
```

Посмотрим на пропуски `living_area` и произведем замену по такому же принципу. `balcony` - имеет почти 50% пропусков. Скорее всего, пропуски говорят о том, что в этих квартирах могло не оказаться балконов и пользователи поставили 0. Здесь прость подставим 0.

Считаю, что как-то правильно заменить оставшиеся колонки нет смысла. Замена не даст должного результата + то что мы пытаемся заменить не требуется для Анализа.

Как мне кажется, пропуски появились из-за того, что не все пользователи заполняли данные. Возможно, в некоторых пунктах что-то плохо выгрузилось. Необходимо передать информацию об этом Дата-инженеру.

Теперь заменим типы данных для дальнейшего удобства.
Что касается замены, то я считаю, нет смысла менять ничего, кроме даты публикации, цены, числа балконов, поскольку все остальные колонки пригодны для дальнейшего анализа.

```python
df = df[np.isfinite(df['floors_total'])]

df['floors_total'] = df['floors_total'].astype(int)
df['last_price'] = df['last_price'].astype(int)
df['balcony'] = df['balcony'].astype(int)
df['is_apartment'] = df['is_apartment'].astype(bool)
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
hist_columns = df.loc[:, ['total_area','last_price', 'ceiling_height', 'price_of_metr', 'rooms']]
for column in hist_columns:
   
    qr = df[column].quantile(.75) - df[column].quantile(.25) #межквартальный размах
   
    mini = df[column].min() # минимальный диапозон
    qmin = df[column].quantile(.25) - qr * 3
    if qmin < mini:
        qmin = mini
    else:
        qmin = qmin
   
    maxi = df[column].max() # максимальный диапазон
    qmax = df[column].quantile(.75) + qr * 3
    if qmax > maxi:
        qmax = maxi
    else:
        qmax = qmax
       
    hist_columns.hist(column, bins = 20, range = ((qmin), (qmax))) #строит гистограммы в нужном диапазоне
```
