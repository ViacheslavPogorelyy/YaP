Как мы видим, cреднее и медиана находятся где-то между 2.7. Если обращать внимание на строительные нормы на 2021 год, то как раз высота потолков 2.7 является железным стандартом. 

Мне кажется что можно прикинуть высоту потолков в зависимости от этажности и от этого произвести замену.

```python
def howlevel(num):
    if num >= 1 and num <= 5:
        return 'малоэтажки'
    elif num > 5 and num < 10:
        return 'многоэтажки'
    else:
        return 'высокоэтажки'

df['new_floor'] = df['floors_total'].apply(howlevel)
```

```python
k = []
for i in ['малоэтажки', 'многоэтажки', 'высокоэтажки']:
    result = df[df['new_floor'] == i]['ceiling_height'].median()
    k.append(result)

level = pd.DataFrame(data = k, columns = ['level'])
```

```python
level
```

Теперь произведем замену пропусков в зависимости от этажности

```python
columns = ['малоэтажки', 'многоэтажки', 'высокоэтажки']
for i in range(len(columns)):
    df.loc[df['new_floor'] == columns[i],'ceiling_height'] = df.loc[df['new_floor'] == columns[i],'ceiling_height'].fillna(level['level'][i])
```

Теперь посмотрим, все ли удалось изменить. 

```python
df['ceiling_height'].isna().sum()
```

Посмотрим на пропуски `living_area` и произведем замену по такому же принципу.
```python
hist_columns = df.loc[:, ['total_area','last_price', 'ceiling_height', 'price_of_metr', 'rooms']]
for column in hist_columns:
   
    qr = df[column].quantile(.75) - df[column].quantile(.25) #межквартальный размах
   
    mini = df[column].min() # минимальный диапозон
    qmin = df[column].quantile(.25) - qr * 3
    if qmin < mini:
        qmin = mini
    else:
        qmin = qmin
   
    maxi = df[column].max() # максимальный диапазон
    qmax = df[column].quantile(.75) + qr * 3
    if qmax > maxi:
        qmax = maxi
    else:
        qmax = qmax
       
    hist_columns.hist(column, bins = 20, range = ((qmin), (qmax))) #строит гистограммы в нужном диапазоне
```

Изучите предложения квартир: для каждой квартиры есть информация о расстоянии до центра. Выделите квартиры в Санкт-Петербурге ('locality_name'). Ваша задача — выяснить, какая область входит в центр. Создайте столбец с расстоянием до центра в километрах: округлите до целых значений. После этого посчитайте среднюю цену для каждого километра. Постройте график: он должен показывать, как цена зависит от удалённости от центра. Определите границу, где график сильно меняется — это и будет центральная зона.

```python
without_trash['nearest_km'] = without_trash['cityCenters_nearest'] / 1000
```

Поскольку у нас остались пропуски в колонке Расстояние до центра, то мы специально заменим их 0, для перевода столбца к целым числам. Для дальнейшено рассчета все нулевые значения устраним. 

```python
without_trash['nearest_km'] = without_trash['nearest_km'].fillna(9999)
```

```python
without_trash['nearest_km'] = (round(without_trash['nearest_km'], 0)).astype(int)
```

```python
spb = without_trash.query('locality_name == "Санкт-Петербург" and nearest_km != 9999').reset_index()
spb = spb.drop(['level_0', 'index'], axis = 1)
```
